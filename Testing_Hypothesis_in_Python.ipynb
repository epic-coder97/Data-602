## Q) Create a sub-sample of this population with mean of 1.5, a sigma of 2.0, and 200 observations.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
mu, sigma = 3, 2
s = np.random.normal(mu, sigma, 10000)
plt.hist(s, bins=25, density=True, alpha=0.5, color='b')

xmin , xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p , 'k', linewidth = 2)

sample_mean , sample_sigma = 1.5, 2
sample = np.random.normal(sample_mean, sample_sigma, 200)
plt.hist(sample, bins =30, density= True, alpha=0.5)
xmin1 , xmax1 = plt.xlim()
x1 = np.linspace(xmin1, xmax1, 100)
p1 = norm.pdf(x1, sample_mean, sample_sigma)
plt.plot(x1, p1 , 'k', linewidth = 2)
plt.show()

```

## Q) Create a graph that shows the null and hypotheses, the type 1 and 2 errors that looks like this:

image.png

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import scipy

from scipy.stats import norm
mu, sigma = 3, 2
s_1 = np.random.normal(mu, sigma, 10000)
_, bins, _ = plt.hist(s_1,30, density=1, alpha=0.1, color='b')
mu, sigma = scipy.stats.norm.fit(s_1)
bfl = scipy.stats.norm.pdf(bins,mu,sigma)
plt.plot(bins, bfl, color = 'b')
sample_mean , sample_sigma = 1.5, 2
s = np.random.normal(sample_mean, sample_sigma, 200)

count, bins, ignored = plt.hist(s, 30, alpha=0.1, density=True, color ='r', range =[-6,10])
mu, sigma = stats.norm.fit(s)
bfl=scipy.stats.norm.pdf(bins,mu,sigma)
plt.plot(bins,bfl,color='r')

from numpy import sqrt
ci = stats.norm.interval(0.95, loc=mu, scale=sigma)
plt.axvline(ci[0],color='g')
plt.axvline(ci[1],color='g')

x = np.arange(-6,ci[0], 0.01)
y1 = stats.norm(mu,sigma).pdf(x)
plt.fill_between(x = x, y1= y1, facecolor='red', alpha = 0.5)

x_2 = np.arange(-6,ci[0], 0.01)
y2 = stats.norm(mu,sigma).pdf(x)
plt.fill_between(x = x_2, y1= y2, facecolor='red', alpha = 0.5)

x_3 = np.arange(-6,ci[0], 0.01)
y3 = stats.norm(mu,sigma).pdf(x)
plt.fill_between(x = x_3, y1= y3, facecolor='Blue', alpha = 0.3)

plt.text(x = 0.35 , y= 0.15, s="Null Hypothesis")
plt.text(x = 5.8 , y= 0.03, s="Alternate")

plt.text(x = 5.2 , y= 0.002, s="Type1 Error")
plt.text(x = 4.5 , y= 0.005, s="Type1 Error")
plt.text(x = 1.5 , y= 0.1, s="Type2 Error")
plt.show()

```

```python
z_score=(sample_mean-mu)/sigma
p_value = scipy.stats.norm.sf(abs(z_score))
print('P-value= {}'.format(p_value))
if p_value<0.05:
    print('P-value<alpha: reject H0')
else:
    print('P-value > alpha: do not reject H0')
```
